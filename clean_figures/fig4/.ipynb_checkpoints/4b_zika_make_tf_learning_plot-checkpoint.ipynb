```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from glob import glob

import matplotlib
import matplotlib.font_manager as fm
import os
import re
import scipy.stats as sp
#fpath = '/Library/Fonts/Myriad-Pro-Light_31642.ttf'
fpath = '/Library/Fonts/Myriad-Pro_31655.ttf'
prop = fm.FontProperties(fname=fpath)
fname = os.path.split(fpath)[1]
axis_font = {'fontname':fname, 'fontproperties':prop}

#matplotlib.rcParams['font.family'] = "sans-serif"
#matplotlib.rcParams['font.sans-serif'] = "Myriad Pro"
```

# Part 1: Load in data

```python
in_dir = 'cleaned_csvs/'
df = pd.read_csv(in_dir + '4b_zika_tf_learning_metrics.csv')

df
```

# Part 2: Make plots

```python
blue = '#407DCF'
grey = 'lightgray'

x = np.arange(len(df))  # the x locations for the groups
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize = (3,3), dpi = 300)

ax.bar(x, df['spearman'], width,color = blue,edgecolor = 'black', alpha = 0.7)
ax.plot(x, df['spearman'],marker = 'o', linestyle = '', color = blue, markeredgecolor = 'black',label='Spearman')

ax.plot(x, [0, 0, 0, 0], color='k', linestyle='--', linewidth=1)

ax.set_yticklabels([np.round(x, 2) for x in ax.get_yticks()], **axis_font, fontsize = 15)
ax.set_ylabel('Spearman Rank Correlation', **axis_font, fontsize = 15)

ax.set_xticks(x)
ax.set_xticklabels(['Original\nModel', 'Model Trained\non Green Data', 'Transfer:\nInitialize Weights', 'Transfer:\nFreeze Weights'], rotation = 45, **axis_font, fontsize = 12)
ax.set_xlabel('Model', **axis_font, fontsize = 15)

ax.set_title('Predictions on Zika Toeholds', **axis_font, fontsize = 15)

ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#plt.savefig('../../illustrator_files/fig4/fig4_individuals/4b_zika_transfer.svg')
plt.show()
```
