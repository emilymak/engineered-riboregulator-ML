These plots will be Fig 1c and 1d, the GC content and MFE distributions of all switches vs top performing switches.

```python
# import statements
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn
from collections import Counter
import scipy.stats as sp
import logomaker as lm

import matplotlib
import matplotlib.font_manager as fm
import os
fpath = '/Library/Fonts/Myriad-Pro_31655.ttf'
prop = fm.FontProperties(fname=fpath)
fname = os.path.split(fpath)[1]
axis_font = {'fontname':fname, 'fontproperties':prop}
```

# Part 1: Load in and clean up data for use

```python
# load in data
data_dir = '../../data/'
file_name = 'newQC_toehold_data.csv'
data_df = pd.read_csv(data_dir + file_name,sep=',')
print(data_df.head(5))
```

### Filter the data based on QCs, as defined in Angenent-Mari et al.

```python
new_data_df = data_df[~np.isnan(data_df['on_value'])]
new_data_df = new_data_df[~np.isnan(new_data_df['off_value'])]
new_data_df = new_data_df[new_data_df['off_qc'] >= 1.1]
new_data_df = new_data_df[new_data_df['on_qc'] >= 1.1].reset_index()

toehold_seqs = new_data_df['switch_sequence']
seq_len = len(toehold_seqs[0])

print('Number of remaining sequences: ', len(new_data_df))
```

```python
# Extract top performing sequences from the experimental data
sorted_exp = new_data_df.sort_values(by = 'onoff_value')
cutoff = np.percentile(sorted_exp['onoff_value'], 95) # return 95th percentile and above
print(cutoff)
sorted_exp = new_data_df[new_data_df['onoff_value'] > cutoff]
print(sorted_exp.shape)
```

```python
# Extract bottom performing sequences from the experimental data
bottom_sorted_exp = new_data_df.sort_values(by = 'onoff_value')
cutoff = np.percentile(bottom_sorted_exp['onoff_value'], 5) # return 5th percentile and below
print(cutoff)
bottom_sorted_exp = new_data_df[new_data_df['onoff_value'] < cutoff]
print(bottom_sorted_exp.shape)
```

# Part 2: Query GC Content

```python
def query_GC_content(one_seq):
    length = len(one_seq)
    if (length < 1):
        print("Bad length")
    
    gc_tot = 0
    for nt in range(0, length):
        curr_nt = one_seq[nt]
        if (curr_nt == "G" or curr_nt == "C"):
            gc_tot += 1
    
    return gc_tot / float(length)
```

```python
all_seqs = new_data_df.loc[:,'switch_sequence']
gc_conts = [query_GC_content(x) for x in all_seqs]

subset_seqs = sorted_exp.loc[:, 'switch_sequence']
subset_gc_conts = [query_GC_content(x) for x in subset_seqs]

bot_subset_seqs = bottom_sorted_exp.loc[:, 'switch_sequence']
bot_subset_gc_conts = [query_GC_content(x) for x in bot_subset_seqs]
```

```python
fig, ax = plt.subplots(figsize = (15, 10), dpi = 300)
ax.patch.set_facecolor('white')

bins = np.linspace(0.1, 0.75, 40)
weights = np.ones_like(subset_gc_conts)/float(len(subset_gc_conts))
plt.hist(subset_gc_conts, color = 'tab:blue', label='Top 5% Sequences', weights = weights, density=False, bins = bins, alpha = 0.5, edgecolor='black', linewidth=0.3)

weights = np.ones_like(gc_conts)/float(len(gc_conts))
plt.hist(gc_conts, color = 'tab:orange', label='All Sequences', weights = weights, density=False, bins = bins, alpha = 0.5, edgecolor='black', linewidth=0.3)

mwu = sp.mannwhitneyu(gc_conts, subset_gc_conts)
print(mwu)
mwu = sp.mannwhitneyu(gc_conts, bot_subset_gc_conts)
print(mwu)

#removing top and right borders
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
ax.spines['right'].set_color('white')
ax.spines['top'].set_color('white')
legend_prop = fm.FontProperties(fname=fpath, size = 35)
plt.legend(loc='upper right', prop=legend_prop, facecolor = 'white', fancybox = True)

plt.xlabel('GC Content', **axis_font, fontsize = 35)
plt.ylabel('Frequency', **axis_font, fontsize = 35)
ax.set_xticklabels([np.round(x, 2) for x in ax.get_xticks()], **axis_font)
ax.set_yticklabels([np.round(x, 3) for x in ax.get_yticks()], **axis_font)

ax.tick_params(length = 10, width = 1)
plt.xticks(fontsize =30)
plt.yticks(fontsize = 30)
output_dir = '../../illustrator_files/fig1/fig1_individuals/'
#plt.savefig(output_dir + 'sfig3_gc_cont.svg', format='svg')
plt.show()
```

# Part 3: Make MFE Distributions
The new version of the QC dataframe does not have MFE values. We need to go into the old dataframe and extract the downsampled relevant switches / rows to query those.

```python
# load in old data, downsampled and pruned to get additional info
filename = 'downsampled_pruned_old_data.csv'
old_pruned_df = pd.read_csv(data_dir + filename,sep=',')
old_pruned_df.head()
```

```python
switch_mfe = old_pruned_df['switch_mfe']
subset_switches = sorted_exp.loc[:, 'switch_sequence']

tests = list(subset_switches)
switches_to_test = list(old_pruned_df['switch_sequence'])
is_old_data_in_new_data = [True if x in tests else False for x in switches_to_test]
print(sum(is_old_data_in_new_data))

subset_downsampled_pruned_old_data = old_pruned_df[is_old_data_in_new_data]
print(subset_downsampled_pruned_old_data.head(5))

subset_switch_mfe = subset_downsampled_pruned_old_data.loc[:, 'switch_mfe']
```

```python
fig, ax = plt.subplots(figsize = (15, 10), dpi = 300)
bins = np.linspace(-40, -5, 40)
weights = np.ones_like(subset_switch_mfe)/float(len(subset_switch_mfe))
plt.hist(subset_switch_mfe, label='Top 5% Sequences', weights = weights, density=False, bins = bins, alpha = 0.5, edgecolor='black', linewidth=0.3, color = 'tab:blue')
weights = np.ones_like(switch_mfe)/float(len(switch_mfe))
plt.hist(switch_mfe, label='All Sequences', weights = weights, density=False, bins = bins, alpha = 0.5, edgecolor='black', linewidth=0.3, color = 'tab:orange')

plt.legend(loc='upper left', prop = legend_prop, facecolor = 'white', fancybox = True)
mwu = sp.mannwhitneyu(switch_mfe, subset_switch_mfe)
print(mwu)

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
plt.xlabel('Minimum Free Energy', **axis_font, fontsize = 35)
plt.ylabel('Frequency', **axis_font, fontsize = 35)

ax.tick_params(length = 10, width = 1)
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
ax.set_xticklabels([int(x) for x in ax.get_xticks()], **axis_font, fontsize = 30)
ax.set_yticklabels([np.round(x, 3) for x in ax.get_yticks()], **axis_font, fontsize = 30)

#plt.savefig(output_dir + 'sfig3_mfe.svg', format='svg')
plt.show()
```
