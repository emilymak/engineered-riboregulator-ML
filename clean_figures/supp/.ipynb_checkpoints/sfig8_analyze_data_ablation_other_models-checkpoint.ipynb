```python
import numpy as np
import seaborn as sns
import pandas as pd
import scipy.stats as sp
```

# Part 1: Load in data for S8A

```python
full_df = pd.read_csv('data/FigS8A.csv')
full_df
```

# Part 2: Do statistical testing

```python
real_736 = full_df.iloc[1,1:6]
scrambled_736 = full_df.iloc[1,6:11]
```

```python
print(sp.ttest_ind(real_736, scrambled_736))
```
