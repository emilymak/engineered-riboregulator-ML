# Part 1: Load in data

# Part 2: Do statistical testing

```python
real_736 = full_df.iloc[1,1:6]
shuffled_736 = full_df.iloc[1,6:11]
scrambled_736 = full_df.iloc[1,11:16]
```

```python
print(sp.ttest_ind(real_736, shuffled_736))
```
